Step response of the 1st order dynamic system

```python
import numpy as np
import matplotlib.pyplot as plt

K = 3
T = 4

start = 0
stop = 30
increment = 0.1
t = np.arange(start, stop, increment)  # make an array 

y = K*(1-np.exp(-t/T))

plt.plot(t, y)
plt.title('1st order dynamic system')
plt.xlabel('t [s]')
plt.ylabel('y(t)')
plt.grid()
```

## 002. Using an ODE Solver 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Initialization 
K = 3
T = 4
u = 1

tstart = 0
tstop = 30
increment = 0.1
y0 = 0

# Function taht returns dx/dt
def system1order(y, t, K, T, u):
    dydt = (1/T) * (-y + K*u)
    return dydt

# Solve ODE
x = odeint(system1order, y0, t, args=(K, T, u))
print(x)

plt.plot(t, x)
plt.title('1st order dynamic system')
plt.xlabel('t [s]')
plt.ylabel('y(t)')
plt.grid()
```

## 003_Discrete systems

```python
import numpy as np
import matplotlib.pyplot as plt

# Model Parameters 
K = 3
T = 4
a = -1/T
b = K/T

# Simulation Parameters
Ts = 0.1
Tstop = 30
uk = 1 # Step response
yk = 0 # Initial Values
N = int(Tstop/Ts)  # Simulation length
data = []   # data storage array 
data.append(yk)

# Simulation
for k in range(N):
    yk1 = (1 + a*Ts)*yk + Ts*b*uk  # discrete version of the differential equation 
    yk = yk1
    data.append(yk1)

# Plot the simulation results
t = np.arange(0, Tstop+Ts, Ts)

plt.plot(t, data)
plt.title('1st order dynamic system')
plt.xlabel('t [s]')
plt.ylabel('y(t)')
plt.grid()
```
